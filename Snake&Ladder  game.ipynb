# Snake and ladder game using Random module

```python
import random
p1=0
p2=0
snakebit={17:7,54:34,62:19,64:60,87:36,92:73,95:75,98:79}
ladder={1:38,4:14,21:42,9:31,28:84,51:67,72:91,80:99}
def SnakeAndladder(pos):
    dice=random.randint(1,6)
    print("Dice:{}".format(dice))
    pos=pos+dice
    if pos in snakebit:
        print("snake bitten")
        pos=snakebit[pos]
        print("Present position:{}".format(pos))
    elif dice==6:
        print("You dice face is {} so you have another chance".format(dice))
        input("press the key to rethrow ")
        dice_2=random.randint(1,6)
        print("dice face {}".format(dice_2))
        pos+=dice_2
        print("Present position:{}".format(pos))
    elif pos>100:
        print("out of range ")
        pos-=dice
        print("Present position {}".format(pos))
    elif pos in ladder:
        print("Climbed by ladder")
        pos=ladder[pos]
        print("Present position:{}".format(pos))
    else:
        print("Present position:{}".format(pos))
    print("\n")
    return pos
while True:
    Player1=input("enter \"Raghu\" to throw dice :")
    p1=SnakeAndladder(p1)
    if p1==100:
        print("Game over!!!\n {} wins".format(Player1))
        break
    Player2=input("enter \"Rohan\" to throw dice :")
    p2=SnakeAndladder(p2)
    if p2==100:
        print("Game over!!!\n {} wins".format(Player2))
        break    
```
